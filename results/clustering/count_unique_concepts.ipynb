```python
import numpy as np
import os 

n_=[]
for file in os.listdir('./evecs'):
    if '.txt' in file:
        try:
            for line_ in open('./evecs/'+file,'r').readlines():
                n_.append(line_.strip('\n'))
        except:
            print(f"Failed to read {file}")
```

```python
import numpy as np
import os 

n_n=[]
for file in os.listdir('./neurons'):
    if '.txt' in file:
        try:
            for line_ in open('./neurons/'+file,'r').readlines():
                n_n.append(line_.strip('\n'))
        except:
            print(f"Failed to read {file}")
```

```python
n_ = np.asarray(n_, dtype='int')
n_n = np.asarray(n_n, dtype='int')
```

```python
np.mean(n_), np.std(n_)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#plt.plot(n_,n_n)
plt.plot(n_)
plt.plot(n_n)
```

```python
n_+=1
n_n+=1
n_[n_==0]=1
n_n[n_n==0]=1
```

```python
plt.rcParams['figure.figsize']=(10,5)
cmap_name = 'Paired'
cmap = plt.cm.get_cmap(cmap_name)

plt.hist(n_, bins=[1,2,3,4,5,6],alpha=1,color=cmap(np.linspace(0, 1, 1)))
#plt.hist(n_n, alpha=0.5,color=cmap(np.linspace(1, 10, 1)))
plt.ylabel("Directions with Y unique concepts")
plt.xlabel("Number Y of unique concepts")
plt.legend(["Our method"])
```
